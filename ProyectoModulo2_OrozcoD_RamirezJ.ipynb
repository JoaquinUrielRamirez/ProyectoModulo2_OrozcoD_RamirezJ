```python
# Generacin de montaa por medio de fractales utilizando montecralo

fig = plt.figure() # environment de figura de 3d
ax = fig.add_subplot(111, projection="3d")

ax.view_init(azim=-120,elev=45) # orientacin de la vista
ax.dist = 5.0 # distancia de la vista
ax.set_facecolor([.1,0.0,0.0]) # color del piso

n = 8 # nmero de ciclos
dx = -0.7 # parmetro de cambio en x inicial
dy = 0.0 # parmetro de cambio en y inicial
L = 1.5 # lado de area de cada cuadrado
M = 200 # numero de pixeles

def f(Z): # funcin de elevacin
    return np.e**(-np.abs(Z))

x = np.linspace(-L+dx,L+dx,M) # x variable array
y = np.linspace(-L+dy,L+dy,M) # y variable array
X,Y = np.meshgrid(x,y) 
Z = np.zeros(M) # inicio de plano completo
W = np.zeros((M,M)) # inicio de valores de elevacin
C = X + 1j*Y # area del plano complejo

for k in range(1,n+1): # ciclo iterativo
    ZZ = Z**2 + C
    Z = ZZ
    W = f(Z) # mdulos de suma
    
ax.set_xlim(dx-L,dx+L) # set x axis limits
ax.set_zlim(dy-L,dy+L) # set y axis limits
ax.set_zlim(-L,2*L) # set z axis limits
ax.axis("off") # do not plot axes
ax.plot_surface(X, Y, W, rstride=1, cstride=1, cmap="pink") # plot surface as a whole
plt.show() # show plot
```
